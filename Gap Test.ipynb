```python
import numpy as np

n=100
t=np.zeros((n))
x=np.zeros((n))
u=np.zeros((n))

a=7**5
c=0
m=(2**31)-1

x[0]=0.5
u[0]=0.5/m
for i in range(0,n):
    t[i]=i

for i in range(0,n-1):
    x[i+1]=(a*x[i]+c)%m
    u[i+1]=(x[i+1])/m

T=[]
T.append(0)
for i in range (n):
    if 0.4< u[i]<0.6:
        T.append(i+1)

print('len',len(T))   
print(T)      
Z=[]

for i in range(1,len(T)):
    zval=T[i]-T[i-1]-1
    Z.append(zval)
print(Z)

r=10

indexZ=np.zeros((r+1))  # Z frequency

for i in range(r+1):   # we need the index value of each Z
    indexZ[i]=i
    
fre=np.zeros((r+1))   # keeps a count of each j
ObsVal=np.zeros((r+1))   # observed val
for i in range(len(Z)):
    for j in range(r+1):
        if Z[i]==indexZ[j]:
            fre[j]=fre[j]+1

for i in range (r+1):
    ObsVal[i]=fre[i]/len(Z)

P = 0.6-0.4
ExpVal=np.zeros((r+1))
ExpVal[r]=(1-P)**r
for i in range(r):
    ExpVal[i]=P*((1-P)**i)

print("probaility array",ExpVal)

chival = 0
for i in range(r+1):
    chival = chival + (((ObsVal[i] - ExpVal[i]) ** 2) / ExpVal[i])

print("Chi value",chival)

#############Check with table value(chi)############
chitable = float(input("Enter the table chi square value:"))
alpha = float(input("Enter the alpha value:"))
if chitable > chival:
    print("Alpha value is:", alpha)
    print("Calculated Chi value:", chival)
    print("Table chi square value:", chitable)
    print("Do not reject null hypothesis")
    print("Random numbers are iid")
else:
    print("Alpha value is:", alpha)
    print("Calculated Chi value:", chival)
    print("Table chi square value:", chitable)
    print("Random numbers are not iid")
```
