```python
## Exponential Generator 

## Generate random numbers (L'Ecuyer)

import numpy as np 
import matplotlib.pyplot as plt
import math

n = 100
m1 = 2**32 - 209
m2 = 2**32 - 22853

x = np.zeros(n+3)
y = np.zeros(n+3)
U = np.zeros(n)   # Generated random numbers 

x[0] = 15
y[0] = 20

## need to choose numbers for x[1],x[2],y[1] & y[2]

x[1] = 8
y[1] = 15

x[2] = 25
y[2] = 5

for i in range(3,n+3):
    x[i] = (1403580*x[i-2] - 810728*x[i-3]) % m1
    y[i] = (527612*y[i-2] - 1370589*y[i-3]) % m2
    
for i in range(3,n+3):
    if x[i]<=y[i]:
        U[i-3] = (x[i]-y[i]+m1)/(m1+1)
    else:
        U[i-3] = (x[i]-y[i])/(m1+1)

## Generating random variables following Exp(3)
# (using inverse-transform method)

U1 = np.zeros(n)  # 1-U~U(0,1)
X = np.zeros(n)   # Generated random variables
L = 3    # Lambda = 3
exp = np.zeros(n) # holds ln(U)

for i in range(n):
    U1[i] = 1 - U[i]
    exp[i] = np.log(U1[i])
    X[i] = (-1/L)*exp[i]

## plotting a histogram 

plt.figure(1)
plt.hist(X,density=True)
plt.xlabel("Random variables generated")
plt.ylabel("Frequency")
plt.title("Random variables following Exp(3)")
plt.grid()
plt.show()


## plotting a histogram 

plt.figure(2)
plt.hist(X,bins=25)
plt.xlabel("Random variables generated")
plt.ylabel("Frequency")
plt.title("Random variables following Exp(3)")
plt.grid()
plt.show()
```
