```python
import numpy as np
from sklearn.linear_model import LinearRegression   #perform linear and polynomial regression and make predictions accordingly
import matplotlib.pyplot as plt 

# 1 - Provide data

x = np.array([3.7, 3.7, 3.3, 2.3, 4.0, 3.7, 2.7, 2.7, 1.0, 3.3, 3.0, 2.7, 3.0]).reshape((-1, 1)) 
y = np.array([3.7, 3.3, 3.3, 3.3, 4.0, 4.0, 2.3, 3.7, 2.0, 4.0, 3.7, 3.0, 3.3])

#.reshape() on x --> this array must be two-dimensional, 
#it must have one column and as many rows as necessary 
#argument (-1, 1) of .reshape() specifies that.

# 2 - Create a model and fit it

model = LinearRegression()
model.fit(x, y)

# OR

# model = LinearRegression().fit(x, y)   #This completes above 2 steps in 1 

# 3 - Get results

r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")   #returns R^2

print(f"intercept: {model.intercept_}")   # returns b0, is a scalar


print(f"slope: {model.coef_}")     # returns b1, an array


# 3 - Another method 

#provide y as a 2D array

new_model = LinearRegression().fit(x, y.reshape((-1, 1)))
print(f"intercept: {new_model.intercept_}")    # b0 is a 1D array, with a single element
print(f"slope: {new_model.coef_}")     # b1 is a 2D array with a single element


# 4 - Predict response

#y_pred = model.predict(x)
#print(f"predicted response:\n{y_pred}")     # Predicted response/Output - 1D array

# OR 

y_pred = model.intercept_ + model.coef_ * x
print(f"predicted response:\n{y_pred}")       # Here, predicted response is a 2D-array


# use fitted models to calculate the outputs based on new inputs:
x_new = np.arange(5).reshape((-1, 1))  #np.arrange(5) = generate an array with elements from 0-5(including 0, excluding 5)
x_new

y_new = model.predict(x_new)
y_new


plt.figure()
plt.plot(x,y,".r")
plt.plot(x,y_pred,"b")
plt.plot(x_new,y_new,"g")
plt.show()


```
