```python
#L'Ecuyer

import numpy as np
import matplotlib.pyplot as plt

n = 1000
m1 = 2**32 - 209
m2 = 2**32 - 22853

x = np.zeros(n)
y = np.zeros(n)
U = np.zeros(n)

x[0] = 10000
x[1] = 20000
x[2] = 30000

y[0] = 25000
y[1] = 35000
y[2] = 45000

for i in range(0,n-3):
    x[i + 3] = (1403580*x[i+1] - 810728* x[i+2]) % m1
    y[i + 3] = (527612*x[i+1] - 1370589* x[i+2]) % m2


for i in range(n):
    if x[i] <= y[i]:
        U[i]=(x[i]-y[i]+m1)/(m1+1)
    else:
        U[i] = (x[i] - y[i]) / (m1 + 1)
        
print("U(t) = ",U)

plt.figure()
x=np.linspace(0,n,1000)
plt.plot(x,U,".b")
plt.show()

```
