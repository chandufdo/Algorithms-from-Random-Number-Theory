```python
## Combined generator - L'Ecuyer

import numpy as np 
import matplotlib.pyplot as plt

n = 1000 
m1 = 2**32 - 209
m2 = 2**32 - 22853

x = np.zeros(n)
y = np.zeros(n)
U = np.zeros(n)   # Generated random numbers 

x[0] = 0.7
y[0] = 0.1

## need to choose numbers for x[1],x[2],y[1] & y[2]

x[1] = 0.4
y[1] = 0.5

x[2] = 0.9
y[2] = 0.8

for i in range(3,n):
    x[i] = (1403580*x[i-2] - 810728*x[i-3]) % m1
    y[i] = (527612*y[i-2] - 1370589*y[i-3]) % m2
    
for i in range(n):
    if x[i]<=y[i]:
        U[i] = (x[i]-y[i]+m1)/(m1+1)
    else:
        U[i] = (x[i]-y[i])/(m1+1)
#print(U)

## plotting the scatter plot 

plt.figure(1)
X = np.linspace(0,n,1000)
plt.plot(X,U,".k")
plt.ylabel("Random numbers generated")
plt.xlabel("n = 1000")
plt.title("Scatter plot")
plt.show()

## plotting the histogram

plt.figure(2)
plt.hist(U,density=True)
plt.xlabel("Random variables generated")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid()
plt.show()
```
