```python
import numpy as np 


## Generate random numbers using LCG

n = 100
x = np.zeros(n)
x[0] = 1000
a = 16807
c = 0
m = 2147483647
U = np.zeros(n)

for i in range(0, n - 1):
    x[i + 1] = ((a * x[i]) + c) % m

for i in range(n):
    U[i] = x[i] / m
    
k = 100  #no of intervals 

inx = np.zeros((k + 1)) 
iny = np.zeros((k + 1)) 
inx[0] = iny[0] = 0 
inx[k] = iny[k] = 1 

##Generate 2 equal partitions (as in frequency test)
 # for both inx and iny 
    
for i in range(1,k):
    inx[i] = iny[i] = ((1-0)*i)/k

print("Number of Random numbers generated", n) 
print("Number of intervals", k) 

# Generate expected Values 
expVal = n / (2 * (k ** 2)) 
print("Expected value = ", expVal) 

# Generate observed values 
obsVal = np.zeros((k, k))  # is a kxk matrix 

for i in range(0,n,2):
    for j in range(k):
        for k in range(k):
            if inx[j]<=U[i]<=inx[j+1]:
                if iny[k]<=U[i+1]<=iny[k+1]:
                    obsVal[j][k] = obsVal[j][k] + 1 
chival = 0 
for i in range(k):
    for j in range(k):
        chival = chival + ((obsVal[i][j] - expVal)**2)/expVal
        
print("Calculated Chi-square value = ",chival)
chitable = float(input("Table chi-square value = "))
alpha = float(input("Significance level (alpha) = "))

if chival < chitable:
    print("Do not reject null hypothesis")
    print("Random numbers are iid")
else:
    print("Reject null hypothesis")
    print("Random numbers are not iid")

                    
```
